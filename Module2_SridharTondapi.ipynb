## 1. Do the users spend more time on the new landing page than the existing landing page?

### Perform Visual Analysis

```python
# import libraries for data manipulation
import numpy as np
import pandas as pd

# import libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Below is used to plot the figure size as per dimensions

plt.figure(figsize=(15,8))

# Below is the boxplot used with seaborn library on the X-axis as 'time_spent_on_the_page' and 'landing_page' and dataframe 'df'

data_histplot = sns.boxplot(data=df, x ='time_spent_on_the_page', y = 'landing_page' );

# xticks is used to rotate the x label

plt.xticks(rotation= 45, ha='right');

# Below are used to represent the X-axis, Y-axis, Bar  titles respectively

plt.xlabel('Time spent on the page', fontsize=14)
plt.ylabel('Landing page', fontsize=14)
plt.title ('Converted Vs Time spent on the Page', fontsize=14);
plt.grid()
```

####Observations:

1. Time spent on the new page is more than the old page.

### Step 1: Define the null and alternate hypothesis

Ho : Total time spent on old landing page is equal to the time spent on the new landing page.

Ha:  Total Time spent on new landing page is greater than the time spent on the old landing page.

### Step 2: Select Appropriate test

In this , we will perform a two- sample independent t-test as the sample means are to be compared from 2 independent populations and observed that standard deviation is unknown as well.

### Step 3: Decide the significance level

we are suppose to use the significance level as 0.05

alpha = 0.05

### Step 4: Collect and prepare data

```python
# To Fetch the time spent by the user group on the new landing page


df_new_user_group = df[df['landing_page'] =='new'].value_counts().reset_index()

df_new_user_group_val = df_new_user_group['time_spent_on_the_page']

df_new_user_group_std = df_new_user_group_val.std()

print ('The Standard deviation for the time spent by new user group is', round(df_new_user_group_std,2))

# Print space between two prints

print (" ")

# To Fetch the time spent by the user group on the old landing page

df_old_user_group = df[df['landing_page'] =='old'].value_counts().reset_index()

df_old_user_group_val =  df_old_user_group['time_spent_on_the_page']

df_old_user_group_std =  df_old_user_group_val.std()

print ('The Standard deviation for the time spent by old user group is', round(df_old_user_group_std,2))
```

### Step 5: Calculate the p-value

```python
# Import the necessary statistical libaries

from scipy.stats import ttest_ind

# Below is the code to generate the test statistic and the p_value

test_stat, p_value = ttest_ind(df_new_user_group_val, df_old_user_group_val, equal_var=False, alternative='greater')

print ("The test statistic for the test is ", test_stat)

print (" ")

print ("The p_value for the test is", p_value)
```

### Step 6: Compare the p-value with $\alpha$

```python
alpha = 0.05 

if p_value <= alpha:
  print (f'Since the {round(p_value,4)} is less than {alpha}, we reject the NULL Hypothesis')
else:
  print (f'Since the {round(p_value,4)} is greater than {alpha}, we fail to reject the NULL Hypothesis')
```

### Step 7:  Draw inference

From the above test, we can understand that the p_value is less than alpha, so we have enough statistical evidence to prove that Total Time spent on new landing page by users is greater than the time spent on the old landing page.

## 2. Is the conversion rate (the proportion of users who visit the landing page and get converted) for the new page greater than the conversion rate for the old page?


# Defining the Hypothesis:

p_op = proportion of users using the old page

p_np = proportion of users using the new page



1.   Proportion of users visiting the landing page and new page





> H0 : p_op = p_np


2.   Proportion of users visting the landing page and get converted for new page is greater than conversion of old page






> Ha : p_np > p_op 







# Identify the necessary test:

As, we are suppose to examine or test the proportion of the users converstion rate from old page to new page, we will use the 2-sample Z-test.

# Identify the level of significance

As, defined above the level of significance for this test is , alpha = 0.05

```python
# Below are the proportions of users who visit the landing page and converted  for the old and new page respectively.

# From the dataframe fetch the converted to 'YES' and use the landing page as OLD

p_op = df[(df['landing_page'] == 'old') & (df['converted'] == 'yes')].value_counts()
p_op_new = p_op.sum()
print ('The users visting the landing page as old and converted:', p_op_new)
print (" ")

# From the dataframe fetch the converted to 'YES' and use the landing page as NEW

p_np = df[(df['landing_page'] == 'new') & (df['converted'] == 'yes')].value_counts()
p_np_new = p_np.sum()
print ('The users visiting the landing page as new and converted:', p_np_new)
```

```python
# Below is the statistical caluclation for the proprotion of the users visiting the landing page and get converted

from statsmodels.stats.proportion import proportions_ztest

# users in the control group 

cntrl_grp_users = df[df['group'] =='control'].value_counts()
cntrl_grp_users_sum = cntrl_grp_users.sum()


# users in the treatment group group 

trtment_grp_users = df[df['group'] =='treatment'].value_counts()
trtment_grp_users_sum = trtment_grp_users.sum()

#p_value_1 is used to avoid confusion with other p_values

test_stat, p_value = proportions_ztest([p_np_new,p_op_new],[cntrl_grp_users_sum,trtment_grp_users_sum], alternative='larger')

print ( ' The test statistic value for the proportion z-test is {0:0.5f} '.format(test_stat))

print (" ")

print (' The p_value for the proportion z-test is {0:0.5f} ' .format(p_value))
```

```python
alpha = 0.05 

if p_value <= alpha:
  print (f'Since the p-value {round(p_value,5)} is less than {alpha}, we reject the NULL Hypothesis')
else:
  print (f'Since the p-value {round(p_value,5)} is greater than {alpha}, we fail to reject the NULL Hypothesis')
```

# Draw Inference

As, the p-value is less than the level of significance ,  reject the NULL Hypothesis.

 we conclude that  the proportion of users who visit the landing page and get converted) for the new page is greater than the conversion rate for the old page.

# **Shapiro-Wilk’s test**

```python
alpha = 0.05 

if shapiro_test.pvalue <= alpha:
  print (f'Since the p-value {round(shapiro_test.pvalue,5)} is less than {alpha}, we reject the NULL Hypothesis')
else:
  print (f'Since the p-value {round(shapiro_test.pvalue,5)} is greater than {alpha}, we fail to reject the NULL Hypothesis')
```

# Draw Inference

It is observed that p-value is greater than 0.05, we fail to reject the NULL hypothesis and response follows a normal distribution.

H0 : The populations of all the groups have equal variance.

Ha : The populations of all the groups donot have equal variance.

```python
# create language specific values (or) data frame of the time spent on the new page by  respective language users 

from scipy.stats import levene


time_spent_English = df[(df['landing_page'] == 'new') & (df['language_preferred'] == 'English')]['time_spent_on_the_page']
#print (time_spent_English)

time_spent_French = df[(df['landing_page'] == 'new') & (df['language_preferred'] == 'French')]['time_spent_on_the_page']
#print (time_spent_French)

time_spent_Spanish = df[(df['landing_page'] == 'new') & (df['language_preferred'] == 'Spanish')]['time_spent_on_the_page']
#print (time_spent_Spanish)


LeveneResult_test = levene(time_spent_English, time_spent_Spanish, time_spent_French)
print (LeveneResult_test.pvalue)
```

```python
alpha = 0.05 

if LeveneResult_test.pvalue <= alpha:
  print (f'Since the p-value {round(LeveneResult_test.pvalue,5)} is less than {alpha}, we reject the NULL Hypothesis')
else:
  print (f'Since the p-value {round(LeveneResult_test.pvalue,5)} is greater than {alpha}, we fail to reject the NULL Hypothesis')
```

# Draw Inference

As the p-value is greater than the level of significance, we fail to reject the NULL hypothesis. From this we can infer, the mean time spent by different language users on new page is same.
